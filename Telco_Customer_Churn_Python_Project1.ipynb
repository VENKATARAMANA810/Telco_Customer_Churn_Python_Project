```python
# Creating subplots with contour plots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

axes  = axes.flatten()

# Define the columns you want to plot
columns = ['gender', 'SeniorCitizen', 'Contract']  # Example columns, replace with your desired columns

#iterate over columns plot count plot
for i , col in enumerate(columns):
    sns.countplot(data=df, x=col, ax=axes[i] , hue="Churn")
    axes[i].set_title(f'Count Plot of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```

```python
# Creating subplots with contour plots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

axes  = axes.flatten()

# Define the columns you want to plot
columns = ['StreamingTV', 'StreamingMovies', 'InternetService']  # Example columns, replace with your desired columns

#iterate over columns plot count plot
for i , col in enumerate(columns):
    sns.countplot(data=df, x=col, ax=axes[i]  , hue = "Churn")
    axes[i].set_title(f'Count Plot of {col}')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Count')
    axes[i].tick_params(axis='x', rotation=90)

plt.tight_layout()
plt.show()
```
